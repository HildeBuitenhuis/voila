# Introduction


For this final project we wanted to investigate wheter there is there a relationship between human development index and societal response to covid (eagerness to test and get vaccinated).

To answer this research question we constructed different HDI groups by taking the mean HDI of three countries with a similar human development index. Then, we compared the total number of cases, tests, vaccinations and people totally vaccinated between these different HDI groups.
